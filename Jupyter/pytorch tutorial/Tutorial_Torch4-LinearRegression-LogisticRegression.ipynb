```python
import torch
import torch.nn as nn
import numpy as np
import torchvision
from sklearn import datasets
from torchvision import transforms
from torchvision.datasets import ImageFolder
import matplotlib.pyplot as plt
import matplotlib as mpl
```

# Regresion Lineal

## Recordar Pasos tipicos
1. Dise√±ar modelo (input,output size, forward pass)
2. Construir Loss y Optimizer
3. Training loop:
    * forward pass: computar prediccion y perdida
    * backward pass: gradientes
    * update weights
    

### Paso 0: preparar la data


```python
X_numpy, y_numpy = datasets.make_regression(n_samples=100, n_features=14, noise=20, random_state=1)
X = torch.from_numpy(X_numpy.astype(np.float32))
y = torch.from_numpy(y_numpy.astype(np.float32))
print("y es un vector fila, queremos convertirlo en un vector columna")
print(y.shape)

y = y.view(y.shape[0], 1)

print("Ahora sera un vector columna")
print(y.shape)

print("Recordar definir numero de samples y features, en este caso coincide con las dimensiones de X")
n_samples, n_features = X.shape
print(n_samples, n_features)
```

### Modelo

```python
input_size = n_features #numero de caracteristicas
output_size = 1 # Solo queremos tener 1 valor de la muestra
model = nn.Linear(input_size, output_size) # La  entrada
```

### Loss y Optimizer

```python
learning_rate = 0.01
criterion = nn.MSELoss() #funcion de loss incorporado por pytorch, en este caso MSE
optimizer = torch.optim.SGD(model.parameters(),lr=learning_rate)
```

### Training Loop

```python
num_epochs = 10000
for epoch in range(num_epochs):
    #forward pass and loss
    y_predicted = model(X)
    loss = criterion(y_predicted,y)
    #backward pass, gradientes
    loss.backward()
    
    #update
    optimizer.step()
    
    # VACIAR GRADIENTES, sino se acomulan
    optimizer.zero_grad()
    
    #Informacion cada ciertas epochs
    if (epoch+1)%10==0:
        print(f'epoch: {epoch+1}, loss = {loss.item():.4f}')
```

```python
#GRAFICAR el modelo
predicted = model(X).detach() #desprenderse del tensor anterior para evitar el tracking de gradientes (basicamente se va a un nuevo vector sin requires_grad)
plt.plot(X_numpy, y_numpy, 'ro')
plt.plot(X_numpy, predicted.numpy(), 'b') #Recordar pasar el tensor a formato numpy para poder graficar
plt.show()
```

# Regresion Logistica
Aca se hara clasificacion binaria
## Recordar Pasos tipicos (Pipeline)
1. Dise√±ar modelo (input,output size, forward pass)
2. Construir Loss y Optimizer
3. Training loop:
    * forward pass: computar prediccion y perdida
    * backward pass: gradientes
    * update weights
    
    
Seleccionaremos una funcion de perdida loss diferente

### Paso 0: Prepara la data

```python
from sklearn.preprocessing import StandardScaler #para escalar caracteristicas
from sklearn.model_selection import train_test_split # separar mas facil la data de train y test

bc = datasets.load_breast_cancer() #problema de clasificacion binaria: intentaremos predecir basado en caracteristicas de entrada si alguien tiene o no cancer en este dataset
X, y = bc.data, bc.target

n_samples, n_features = X.shape
print(n_samples, n_features)
print(str(n_samples)+" muestras y "+str(n_features)+" caracteristicas")

#Separamos la data con la funcion train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=1234)

#Debemos preprocesar la data haciendole un sclae
sc = StandardScaler() # Esto ajusta los datos a que tengan media cero y varianza unitaria, se recomienda hacerlo siempre si se quiere hacer una regresion logistica
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test) #aqui solo transform

# Ahora hay que transformalos a tensor los datos y predicciones
X_train = torch.from_numpy(X_train.astype(np.float32))
X_test = torch.from_numpy(X_test.astype(np.float32))
y_train = torch.from_numpy(y_train.astype(np.float32))
y_test = torch.from_numpy(y_test.astype(np.float32))

# Finalmente hay que reshape los tensor de salidas
y_train = y_train.view(y_train.shape[0],1) #Convertimos un vector fila a un vector columna
y_test = y_test.view(y_test.shape[0],1) #Convertimos un vector fila a un vector columna
```

### Model
En el caso de una regresion logistca el modelo se define como
$$f(x) = wx + b $$
donde aplicamos una funcion sigmoid al final

```python
class LogisticRegression(nn.Module):
    def __init__(self, n_input_features):
        super(LogisticRegression, self).__init__()
        self.linear = nn.Linear(n_input_features, 1) # Tama√±o de entrada es n_input_features (30), salida solo 1
        
    def forward(self,x):
        y_predicted = torch.sigmoid(self.linear(x)) #Aqui aplicamos la funcion sigmoid
        return y_predicted

    
model = LogisticRegression(n_features)
        
```

### Loss y optimizer

```python
learning_rate = 0.05
criterion = nn.BCELoss() #Nueva funcion de perdida, en vez de MSE usamos BCE (Binary Crusader Entropy, o entropia cruzada binaria)
#optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

### Training Loop

```python
num_epochs = 10000
for epoch in range (num_epochs):
    # forward pass and loss
    y_predicted = model(X_train)
    loss = criterion(y_predicted, y_train)
    # backward pass
    loss.backward()
    
    # updates
    optimizer.step()
    
    # Limpiar gradientes
    optimizer.zero_grad()
    
    #informacion entre epochs
    if (epoch+1)%100 == 0:
        print(f'epoch: {epoch+1}, loss = {loss.item():.4f}')
    
```

```python

with torch.no_grad(): #SIEMPRE HAY QUE EVITAR EL TRAKING DE LAS GRADIENTES
    y_predicted = model(X_test)
    y_predicted_cls = y_predicted.round()
    acc = y_predicted_cls.eq(y_test).sum() / float(y_test.shape[0]) #Medir la predicision, para cada acierto se suma y 
    ## por el total
    print(f'accuracy: {acc:.4f}')
```

Por lo tanto tuvimos una precission 0.93%, es buena pero si tuneamos los parametros como el learning rate, optimizador o las epocas podriamos mejorar la precision.
